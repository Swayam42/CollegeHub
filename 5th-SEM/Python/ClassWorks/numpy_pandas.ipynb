## Array Attributes

```python
integers=np.array([[1,2,3],[4,5,6]])
print("matrix:",integers)
print("dtype:",integers.dtype)
print("dimension:",integers.ndim)
print("shape:",integers.shape)
print("No Of Elements:",integers.size)
print("",integers.itemsize)
```

```python
floats=np.array([0.000,0.300,0.1,0.5]) #trailing 0s are not printed in numpy
print("matrix:",floats)
print("dtype:",floats.dtype)
print("dimension:",floats.ndim)
print("shape:",floats.shape)
print("No Of Elements:",floats.size)
print("",floats.itemsize)
```

```python
#iterating through a multidimensional array : 
for row in integers:
    for column in row:
        print(column, end=" ")
    print()
```

```python
#flat : converts ndim to 1-d array
for i in integers.flat:
    print(i,end=" ")
```

`Shallow Copy : The array method view returns a new array object with a   view of the original array object data`

`Reshape & resize : both methods are used to change the array dimension`

`reshape()-> returns a view (shallow copy) of the original array with new dimensions, it doesn't modify the original array 
resize()-> modifies the original array`

```python
num=np.arange(1,7)
print(num)
print(num.reshape(2,3))
print(num.resize(2,3))
```

`flatten() is used to flatten a multi-D array into 1-D array`
`It deep copies the original array's data`

```python
grades=np.array([[87,96,70],[100,87,60]])
print("grades:\n",grades)
flattened=grades.flatten()
print("flattened: ",flattened)
```

```python
#now if we change value in flattened wouldn't affect the original grade array
flattened[2]=23; #flattened is a deep copy of grades
print(flattened)
print(grades)
```

```python
grades=np.array([[87,96,70],[100,87,60]])
print("grades:\n",grades)
flattened=grades.ravel()
print("flattened: ",flattened)
```

```python
#now if we change value in flattened then same change is there in orginal array
flattened[2]=23; #flattened is a shallow copy of grades
print(flattened)
print(grades)
```

`The T attribute returns a transpose view(Shallow Copy) of Original Array`

```python
grades=np.array([[87,96,70],[100,87,60]])
print(grades)
print(grades.transpose())
```

`Stacking means combining arrays by adding more columns and more rows`

```python
grades2=np.array([[23,64,75],[67,37,86]])
print("Horizontal Stack:\n",np.hstack((grades,grades2)))
print("Vertical Stack:\n",np.vstack(((grades,grades2))))
```

### DataFrame

```python
import pandas as pd
#DataFrame from dict
grades_dict={'Wally':[87,67,98],'Bally':[23,86,68],'Gally':[86,57,48]}
grades=pd.DataFrame(grades_dict)
print(grades)
```

```python
grades.index=['Test1','Test2','Test3']
print(grades)
```

```python
#for loc we take index in String as input 
grades.loc['Test1']
```

```python
#iloc takes index as int 
grades.iloc[0]
```

```python
grades.loc['Test1':'Test2']
```

```python
grades.iloc[0:2]
```

```python
grades.loc['Test1':'Test2',['Wally','Bally']]
```

```python
grades.iloc[[0,2],0:2]
```

```python
#Boolean Indexing
grades[grades>=60]
```

```python
#at -> used for any index(Eg. str idx)
print(grades.at['Test2','Wally'])
```

```python
#iat -> used for int index only
print(grades.iat[1,0])
```

```python
#.describe
pd.set_option('precision',2)
print(grades.describe())
```

```python
#Transposing the DataFrames with the T attributes
print(grades.T)
```
