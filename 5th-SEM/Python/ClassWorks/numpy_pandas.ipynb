## Array Attributes

```python
integers=np.array([[1,2,3],[4,5,6]])
print("matrix:",integers)
print("dtype:",integers.dtype)
print("dimension:",integers.ndim)
print("shape:",integers.shape)
print("No Of Elements:",integers.size)
print("",integers.itemsize)
```

```python
floats=np.array([0.000,0.300,0.1,0.5]) #trailing 0s are not printed in numpy
print("matrix:",floats)
print("dtype:",floats.dtype)
print("dimension:",floats.ndim)
print("shape:",floats.shape)
print("No Of Elements:",floats.size)
print("",floats.itemsize)
```

```python
#iterating through a multidimensional array : 
for row in integers:
    for column in row:
        print(column, end=" ")
    print()
```

```python
#flat : converts ndim to 1-d array
for i in integers.flat:
    print(i,end=" ")
```

```python
mat1=np.arange(1,21).reshape(4,5)
print(f"Shape: {mat1.shape}\n {mat1}")
mat1=np.arange(1,21).reshape(2,10)
print(f"Shape: {mat1.shape}\n {mat1}")
```

```python
numbers=np.arange(1,6)
print(numbers)
print(numbers*3) #broadcasting happens to the scalar values performing the operation in each element
print(numbers**2)
```

```python
#Arithmetic Operations
num=np.linspace(1.1,5.5,5)
print(num)
print(num*numbers)
```

```python
#Comparing Arrays:
print(numbers)
print(numbers>=13)
print(num)
print(num<numbers)
print(num==numbers)
```
