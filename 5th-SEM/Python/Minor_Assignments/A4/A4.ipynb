1. Write a Python program to create a list of size N and store the random values in it and find the sum
and average.

```python
import random
n=int(input("Enter n: "))
l=[]
for i in range(0,n):
    l.append(random.randint(1,10))
print(l)
s=sum(l)
avg=s/len(l)
print(f"sum: {s}")
print(f"Avg: {avg}")
```

3. Design and develop a menu-driven Python program for the following list operations.  
a. Create a list of N integers  
b. Display the list elements  
c. Insert an element at a specific position  
d. Delete an element at a given position  
e. Exit  

```python
l=[]
c=True
while c:
    print("""1.Create a list of N integers
             2. Display the list elements
             3. Insert an element at a specific position
             4. Delete an element at a given position
             5. Exit""")
    choice=int(input("Enter choice: "))
    match choice:
        case 1:
            l.clear()
            n=int(input('Enter size of list: '))
            for i in range(n):
                e=int(input("Enter element: "))
                l.append(e)
        case 2:
            print(l)
        case 3:
            pos=int(input("Enter position:"))
            u=int(input("Enter the element"))
            l.insert(pos,u)
        case 4:
            i=int(input("Enter the index for deletion:"))
            if i>=0 and i<n:
                del l[i]
            else:
                print("Index out of bounds")
        case 5:
            c=False
        case _:
            print("Invalid Choice")
            
            
            
            
            
```

6. Input 10 integers from the keyboard into a list. The number to be searched is entered through the
keyboard by the user. Write a Python program to find if the number to be searched is present in the
list and if it is present, display the number of times it appears in the list.

```python
l = []
print("Enter 10 integers:")
for i in range(10):
    num = int(input(f"Enter number {i+1}: "))
    l.append(num)

search = int(input("\nEnter the number to search: "))

if search in l:
    count = l.count(search)
    print(f"{search} is present {count} time(s) in the list.")
else:
    print(f"{search} is not present in the list.")
```

7. Write a function that takes a list of numbers as input from the user and produces the corresponding
cumulative list where each element in the list at index i is the sum of elements at index j ≤ i.

```python

def cumulative_list():
    n = int(input("How many numbers? "))
    lst = []
    print("Enter the numbers:")
    for i in range(n):
        num = int(input())
        lst.append(num)
    
    cum = []
    total = 0
    for x in lst:
        total = total + x
        cum.append(total)
    
    print("Original list :", lst)
    print("Cumulative list:", cum)
cumulative_list()
```

8. Write a function that takes n as an input and creates a list of n lists such that ith list contains the first
five multiples of i.

```python
# Question 8
n = int(input("Enter value of n: "))

result = []
for i in range(1, n+1):           # i goes from 1 to n
    multi = []                    # empty list for multiples of i
    for j in range(1, 6):         # first 5 multiples
        multi.append(i * j)
    result.append(multi)

for i in range(len(result)):
    print(f"Multiples of {i+1}: {result[i]}")
```
