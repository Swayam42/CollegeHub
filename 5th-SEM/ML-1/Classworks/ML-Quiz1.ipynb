```python
import numpy as np
arr = []
for i in range(5, 26):
    arr.append(i)
arr = np.array(arr)
even = arr[arr % 2 == 0]
mean = np.mean(even)
var = np.var(even)
std = np.std(even)
print(f"Even : {even}\nMean: {mean}\nVar: {var}\nStd Deviation: {std}")
```

```python
import pandas as pd
data = {
    'Student': ['Ram', 'Sham', 'Kartik', 'Ashok'],
    'Age': [20, 21, 19, 22],
    'Marks': [85, 12, 80, 45]
}
df = pd.DataFrame(data)
df['Grade'] = ['Pass' if m >= 40 else 'Fail' for m in df['Marks']]
print(df)
print(df['Grade'].value_counts())
```

```python
import numpy as np
import matplotlib.pyplot as plt
values = np.random.uniform(0,1,500)
plt.hist(values)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of 500 Random Values')
plt.show()
mean = np.mean(values)
std = np.std(values)
print(f"Mean: {mean:.3f}, Standard Deviation: {std:.3f}")
```
