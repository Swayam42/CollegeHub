```python
import matplotlib.pyplot as plt
X=range(-15,16)
Y=[x**2 for x in X]
plt.plot(X,Y,color='r')
plt.xlabel("X-Axis Points")
plt.ylabel("Y-Axis Points")
plt.title("Plotting y=x^2")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
X=range(-15,16)
Y=[x**2 for x in X]
plt.plot(X,Y,color='r',marker='o',markersize=15)
plt.xlabel("X-Axis Points")
plt.ylabel("Y-Axis Points")
plt.title("Plotting y=x^2")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)
X=np.random.rand(50)
Y=np.random.rand(50)
plt.scatter(X,Y,color='cyan',marker='o',s=100,edgecolors='black',alpha=0.3,label='Data PO')
plt.xlabel("X-Axis Points")
plt.ylabel("Y-Axis Points")
plt.title("Plotting")
plt.grid(True)
plt.show()
```
