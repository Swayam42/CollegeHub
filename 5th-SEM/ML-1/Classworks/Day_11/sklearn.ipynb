```python
from sklearn.tree import DecisionTreeClassifier as dtc
predictors=data.iloc[:,0:9]
predictors.head()
```

```python
target=data.iloc[:,9]
target.head()
```

```python
from sklearn.model_selection import train_test_split as tts
predictors_train, predictors_test, target_train, target_test=tts(predictors, target, test_size=0.3,random_state=123)
predictors_train.head()
```

```python
dtree_entropy =dtc(criterion="entropy",random_state=100,max_depth=3,min_samples_leaf=5)
model=dtree_entropy.fit(predictors_train,target_train)
prediction=model.predict(predictors_test)
print(prediction)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(target_test, prediction, normalize=True))
```
