Q5 : ECLAT algorithm
transactions are given as :
    T1={A,B,C}
    T2={A,C}
    T3={A,B}
    T4={A,C}
    Convert transactions into TID lists
    compute support for the itemsets (A,B),(A,C) and (B,C)
    identify frequent for minimum support =2

```python
transactions = [
    {"A","B","C"},
    {"A","C"},
    {"A","B"},
    {"A","C"}
]

items = sorted(set().union(*transactions))
tid = {i: {t for t, s in enumerate(transactions, start=1) if i in s} for i in items}

pairs = [("A","B"), ("A","C"), ("B","C")]
support = {p: len(tid[p[0]] & tid[p[1]]) for p in pairs}

min_sup = 2
frequent = {p: support[p] for p in support if support[p] >= min_sup}

print(tid)
print(support)
print(frequent)
```
