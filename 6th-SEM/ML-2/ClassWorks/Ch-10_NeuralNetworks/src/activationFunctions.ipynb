```python
def biological_neuron(signals,threshold=1.0):
    soma=sum(signals)
    if soma >=threshold:
        return 1  #neuron fires
    return 0
signals=[0.3,0.4,0.1]
print("Neuron output:",biological_neuron(signals))
```

```python
#Activation Functions
import math
def identity(x):
    return x

def step(x):
    return 1 if x>=0 else 0

def threshold(x,theta=0.5):
    return 1 if x>=theta else 0

def relu(x):
    return max(0,x)

def sigmoid(x,k=1):
    return 1/(1+math.exp(-k*x))

def bipolar_sigmoid(x):
    return (2/(1+math.exp(-x)))-1

#hyperbolic tanget
def tan_h(x):
    return math.tanh(x)
```

```python
# Input and Weights
x = np.array([0.5, 0.2, 0.1])
w = np.array([0.4, -0.7, 0.2])

# Weighted Sum
y_sum = np.dot(x, w)

print("Input Vector :", x)
print("Weight Vector:", w)
print("Weighted Sum :", y_sum)
print()

# Applying all activation functions in order
print("Identity Output        :", identity(y_sum))
print("Step Output            :", step(y_sum))
print("Threshold Output       :", threshold(y_sum))
print("ReLU Output            :", relu(y_sum))
print("Sigmoid Output         :", sigmoid(y_sum))
print("Bipolar Sigmoid Output :", bipolar_sigmoid(y_sum))
print("Tanh Output            :", tan_h(y_sum))
```

```python
import numpy as np
def artificial_neuron(inputs,weights,bias=0):
    y_sum=np.dot(inputs,weights)+bias
    return y_sum
x=np.array([1,0,1])
w=np.array([0.5,-0.6,0.2])
print("SummationOutput: ",artificial_neuron(x,w,bias=0.1))
```
