```python
transactions = [
    {"Bread", "Milk", "Egg", "Butter", "Salt", "Apple"},
    {"Bread", "Milk", "Egg", "Apple"},
    {"Bread", "Milk", "Butter", "Apple"},
    {"Milk", "Egg", "Butter", "Apple"},
    {"Bread", "Egg", "Salt"},
    {"Bread", "Milk", "Egg", "Apple"}
]
print(transactions)
```

```python
import pandas as pd

# ----------------------------------------------------
# STEP 1: Get all unique items from all transactions
# ----------------------------------------------------
# set().union(*transactions)
# - set() creates an empty set
# - union(*transactions) combines all items from all transactions
# - set removes duplicate items automatically

all_items = set().union(*transactions)

# sorted() arranges the items in alphabetical order
items = sorted(all_items)

# ----------------------------------------------------
# STEP 2: Create a list to store row data
# ----------------------------------------------------

data = []   # empty list to store each transaction as a dictionary

# ----------------------------------------------------
# STEP 3: Convert each transaction into 0 and 1 format
# ----------------------------------------------------

for t in transactions:
    row = {}   # create empty dictionary for one transaction
    
    for item in items:
        # If the item is present in transaction → put 1
        # If not present → put 0
        if item in t:
            row[item] = 1
        else:
            row[item] = 0
    
    # Add this row dictionary to the data list
    data.append(row)

# Print the list of dictionaries
print(data)

# ----------------------------------------------------
# STEP 4: Convert list of dictionaries into DataFrame
# ----------------------------------------------------

df = pd.DataFrame(data)

# ----------------------------------------------------
# STEP 5: Set transaction numbers as index (1,2,3...)
# ----------------------------------------------------

df.index = range(1, len(df) + 1)

# Rename the index column
df.index.name = "Transaction"

# Print final DataFrame
print(df)
```

```python
# ---------------------------------------------
# Function to calculate SUPPORT
# ---------------------------------------------
def support(itemset, transactions):
    """
    Support = (Number of transactions containing the itemset)
              ------------------------------------------------
                   (Total number of transactions)
    """
    
    count = 0  # to count how many transactions contain the itemset
    
    for t in transactions:
        # issubset() checks if all items in itemset are inside transaction t
        if itemset.issubset(t):
            count += 1
    
    # divide by total number of transactions
    return count / len(transactions)


# ---------------------------------------------
# Function to calculate CONFIDENCE
# ---------------------------------------------
def confidence(X, Y, transactions):
    """
    Confidence(X -> Y) = Support(X ∪ Y) / Support(X)
    """
    # X.union(Y) combines both itemsets
    return support(X.union(Y), transactions) / support(X, transactions)

# Support of Bread
print("Support(Bread):", support({"Bread"}, transactions))

# Confidence of Bread -> Milk
print("Confidence({Bread} -> {Milk}):",
      confidence({"Bread"}, {"Milk"}, transactions))

# Confidence of {Bread, Milk} -> {Egg}
print("Confidence({Bread, Milk} -> {Egg}):",
      confidence({"Bread", "Milk"}, {"Egg"}, transactions))
```

```python
# Minimum Support Threshold

min_support = 0.5


# Step 1: Find Frequent 1-itemsets (L1)

# Get all unique items
items = set().union(*transactions)

# Keep items whose support >= min_support
freq_1 = []

for i in items:
    if support({i}, transactions) >= min_support:
        freq_1.append({i})

print("Frequent 1-itemsets (L1):")
for i in freq_1:
    print(i, "Support =", support(i, transactions))


# Step 2: Generate Candidate 2-itemsets (C2)

from itertools import combinations

# Convert sets to single items
single_items = [list(i)[0] for i in freq_1]

# Generate all 2-item combinations
candidates_2 = [set(c) for c in combinations(single_items, 2)]


# Step 3: Apply Apriori Pruning to get L2

freq_2 = []

for c in candidates_2:
    if support(c, transactions) >= min_support:
        freq_2.append(c)

print("\nFrequent 2-itemsets (L2):")
for c in freq_2:
    print(c, "Support =", support(c, transactions))
```
