```python
#Instance Methods
class Student:

    def setName(self,name): #Mutator Method
        self.name=name

    def getName(self): #Accessor Method
        return self.name
    
    def setMarks(self,marks):
        self.marks=marks

    def getMarks(self):
        return self.marks

n=int(input("Enter no. of Students: "))
for i in range(n):
    s=Student()
    name=input(f"Enter name of Student {i}: ")
    s.setName(name)
    marks=int(input("Enter marks: "))
    s.setMarks(marks)
    print(f"Hi {s.getName()}, your marks are {s.getMarks()}")
    
```

```python
#Class Methods
class Car:
    wheels=4
    def __init__(self,make,model):
        self.make=make
        self.model=model
    
    @classmethod
    def set_wheels(cls,count):
        cls.wheels=count
    
    def display(self):
        return f"Car: {self.make} {self.model} wheels: {self.wheels}"
    
Car.set_wheels(3) 
car1=Car('A','X')
print(car1.display())
car2=Car('B','Y')
print(car2.display())
```

```python
class MyClass:
    n=0
    def __init__(self):
        MyClass.n=MyClass.n+1
    
    @staticmethod
    def noobjects():
        print("No of instances created: ",MyClass.n)

obj1=MyClass()
obj2=MyClass()
obj3=MyClass()
MyClass.noobjects()
```

```python
# Encapsulation

class Student:
    def __init__(self,id,name):
        self.__id=id      # private variable
        self.__name=name  # private variable
    
    def get_id(self):
        return self.__id

    def set_id(self, id):
        self.__id=id

    def get_name(self):
        return self.__name

    def set_name(self, name):
        self.__name = name

    def display(self):
        print("Hi", self.__name)
        print("ID:", self.__id)

def main():
    obj = Student(1, "Nalla")
    obj.display()
    obj.set_id(12)#encapsulated
    obj.display()

if __name__=="__main__":
    main()

```
