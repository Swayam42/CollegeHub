```python
class Student:
    
    #default constructor(takes only 1 parameter: self) - direct initialisation
    def __init__(self):
        self.name="rohan"
        self.age=20
        self.marks=100
        
    def talk(self):
        print(f"Hello, I am {self.name}")
        print(f"My age is {self.age}")
        print(f"My marks are {self.marks}")
        
def main():
    s1=Student()
    s1.talk()
if __name__=="__main__":
    main()
```

```python
class Student:
    
    #Parameterized Constructor (more than 1 parameter)
    def __init__(self,name,age,marks):
        self.name=name
        self.age=age
        self.marks=marks
        
    def talk(self):
        print(f"Hello, I am {self.name}")
        print(f"My age is {self.age}")
        print(f"My marks are {self.marks}")
        
def main():
    s1=Student("Aalu",100,-24)
    s1.talk()
if __name__=="__main__":
    main()
```

```python
class Student:
    
    #Parameterized Constructor - initialising the value in parameter 
    def __init__(self,name="Ashok"):
        self.name=name
        
    def talk(self):
        print(f"Hello, I am {self.name}")
        
def main():
    s1=Student() 
    s1.talk()
if __name__=="__main__":
    main()
```

```python
class Sample:
    def __init__(self):
        self.x=10 #Instance Variable
    
    def modify(self):
        self.x+=1
def main():
    s1=Sample()
    s2=Sample()
    print('x in s1=',s1.x)
    print('x in s2=',s2.x)
    s1.modify()
    print('x in s1=',s1.x)
    print('x in s2=',s2.x)
if __name__=="__main__":
    main()
```

```python
class Sample:
    x=10 #Class/Static Variable
    @classmethod
    def modify(cls):
        cls.x+=1
        
def main():
    s1=Sample()
    s2=Sample()
    print('x in s1=',s1.x)
    print('x in s2=',s2.x)
    s1.modify()
    print('x in s1=',s1.x)
    print('x in s2=',s2.x)
if __name__=="__main__":
    main()
```
