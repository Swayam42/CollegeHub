```python
from nltk.corpus import stopwords
from pathlib import Path
from textblob import TextBlob
import pandas as pd
from matplotlib import pyplot as plt

stop_words = stopwords.words('english')

text = Path(r'files/test.txt').read_text(encoding='utf-8')

blob = TextBlob(text)

items = blob.word_counts.items()
print(items)
```

```python
items=[i for i in items if i[0] not in stop_words]
print(items)
```

```python
from operator import itemgetter
sorted_items=sorted(items,key=itemgetter(1),reverse=True)
print(sorted_items)
```

```python
top20=sorted_items[0:20]
print(top20)
```

```python
df=pd.DataFrame(top20,columns=['Word','Count'])
print(df)
```

```python
axes=df.plot.bar(x='Word',y='Count',legend=False)
plt.gcf().tight_layout()
plt.show()
```

```python
import spacy
nlp=spacy.load('en_core_web_sm')
document=nlp('In 1994, Tim Berners-Lee founded the.'+'World Wide Web Consortium(W3C), devoted developing web technology')
for entity in document.ents:
    print(f'{entity.text}:{entity.label_}')
```
