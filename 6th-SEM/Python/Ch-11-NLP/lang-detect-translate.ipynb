```python
from textblob import TextBlob
from textblob.sentiments import NaiveBayesAnalyzer
text="Today is a beatiful day.Tomorrow looks like bad weather"
blob=TextBlob(text,analyzer=NaiveBayesAnalyzer())
for sentence in blob.sentences:
    sentiment=sentence.sentiment
    print(f"Sentence : {sentence}")
    print(f"Classification: {sentiment.classification}")
    print(f"Positive Probability : {sentiment.p_pos:2f}")
    print(f"Negative Probability : {sentiment.p_pos:2f}")
```

# Stemming and Lemmatization

- **Stemming** and **lemmatization** are common preprocessing operations in Natural Language Processing (NLP).  
- Both aim to reduce a word to its **base form**.  

### Stemming
- Removes prefixes or suffixes from a word, leaving a **stem**.  
- The stem **may not be a real word**.  
- **Example:** `"running"` → `"run"`, `"better"` → `"better"` (stem may be non-dictionary form).

### Lemmatization
- Reduces a word to its **dictionary form (lemma)**.  
- The result is always a **real word**.  
- **Example:** `"running"` → `"run"`, `"better"` → `"good"`.

```python
!pip install spacy
import spacy
from pathlib import Path

nlp = spacy.load('en_core_web_sm')

file1_path = Path(r'files\doc1.txt')
file2_path = Path(r'files\doc2.txt')

text1 = file1_path.read_text(encoding='utf-8')
text2 = file2_path.read_text(encoding='utf-8')

doc1 = nlp(text1)
doc2 = nlp(text2)

similarity_score = doc1.similarity(doc2)
print(f"Similarity between the two text files: {similarity_score:.2f}")
```

```python
from textblob import Word
from textblob.wordnet import VERB, NOUN, ADJ, ADV
word=Word("run")
verb_synsets=word.get_synsets(pos=VERB)
print("Verb Synsets: ")
for synset in verb_synsets:
    print(f"{synset.name()}:{synset.definition()}")

noun_synsets=word.get_synsets(pos=NOUN)
print("\nNoun Synsets:")
for synset in noun_synsets:
    print(f"{synset.name()}:{synset.definition()}")    
```

```python
from textblob import Word
word=Word("happy")
for i in word.synsets:
    for lemma in i.lemmas():
        print(lemma)
```

```python
from textblob import Word
synonyms=set()
word=Word('happy')
for i in word.synsets:
    for lemma in i.lemmas():
        synonyms.add(lemma.name())
print(synonyms)
```
