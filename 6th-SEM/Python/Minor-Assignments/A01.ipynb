```python
class BankAccount:
    
    def __init__(self,name,accId,balance):
        self.name=name
        self.accId=accId
        self.balance=balance
        
    def withdraw(self,amount):
        if(amount>self.balance):
            print("Gareeb")
        else:
            self.balance-=amount
            print(f"Withdrawn :{amount} \nBalance: {self.balance}")
            
    def deposit(self,amount):
        self.balance+=amount
        print(f"Deposited :{amount} to your account\nBalance: {self.balance}")
        
def main():
    boi=BankAccount("Chinmaya",123,45)
    boi.deposit(5000)
    boi.withdraw(30)
    
if __name__=="__main__":
    main()
```

```python
class Chapter:
    def __init__(self,title,pageCount):
        self.title=title
        self.pageCount=pageCount

class Book:
    def __init__(self,bookName):
            self.bookName=bookName
            self.chapters=[]
            
    def addChapter(self,chapter):
        self.chapters.append(chapter)
    
    def noOfPages(self):
        totalPages=0
        for i in self.chapters:
            totalPages+=i.pageCount
        return totalPages
        
    
def main():
    B1=Book("Physics Book")
    c1=Chapter("Motion",45)
    c2=Chapter("Thermodynamics",34)
    c3=Chapter("Light",24)
    B1.addChapter(c1)
    B1.addChapter(c2)
    B1.addChapter(c3)
    
    print(B1.noOfPages())
            

if __name__ =="__main__":
    main()
```
