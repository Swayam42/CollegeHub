```python
import pandas as pd
import matplotlib.pyplot as plt
celsius=lambda f:5/9*(f-32)
temps =[(f,celsius(f))for f in range(0,101,10)]
temps_df=pd.DataFrame(temps,columns=['Fahreinheit','Celsius'])
axes=temps_df.plot(x="Fahreinheit",y="Celsius",style="-",legend=True)
axes.set_ylabel("Celsius")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[0,2,4,6,8]
y=[0,4,16,36,64]
fig,ax=plt.subplots()
ax.plot(x,y,marker='o',label='Data Points')
ax.set_title("Basic Components of Matplotlib Figure")
ax.set_xlabel("X-Axis")
ax.set_ylabel("Y-Axis")
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
movies=["Annie Hall","Ben-Hur","Casablanca","Gandhi","West Side Story"]
num_oscars=[5,11,3,8,10]
plt.bar(range(len(movies)),num_oscars)
plt.title("My Fav movies")
plt.ylabel("# of Academy Awards")
plt.xticks(range(len(movies)),movies)
plt.show()
```

```python
from collections import Counter
from matplotlib import pyplot as plt

grades=[83,96,56,75,34,90,0,32,75,100,55]
histogram =Counter(min(grade//10*10,90)for grade in grades)
plt.bar([x+5 for x in histogram.keys()],
       histogram.values(),
       10,
       edgecolor=(0,0,0),color="#FFFF00")
plt.axis([-5,105,0,5])
plt.xticks([10*i for i in range(11)])
plt.xlabel("Decile")
plt.ylabel("No of Students")
plt.show()
```
